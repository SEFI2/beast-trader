```python
def retrive_signal_supertrend(df, length=7, multiplier=3):
    multiplier = float(multiplier)
    sti = ta.supertrend(df['high'], df['low'], df['close'], length, multiplier)
    trend_dir = f"SUPERTd_{length}_{multiplier}"
    df["buy"] = (sti[trend_dir] > 0) & (sti[trend_dir].shift() < 0)
    df["sell"] = (sti[trend_dir].shift() > 0) & (sti[trend_dir] < 0)
    return df
retrive_signal_supertrend(df)
```
